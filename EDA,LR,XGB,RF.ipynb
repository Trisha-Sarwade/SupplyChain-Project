```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

# Load data
#x, y = load_data()

# Define XGBoost Regressor
xgb_reg = xgb.XGBRegressor()

# Define hyperparameters for Grid Search
parameters = {'n_estimators': [100, 500, 1000],
              'learning_rate': [0.01, 0.1, 0.5],
              'max_depth': [3, 6, 9]}

# Create Grid Search object
grid_search = GridSearchCV(estimator=xgb_reg, param_grid=parameters,
                           cv=5, n_jobs=-1, scoring='neg_mean_squared_error')

# Fit Grid Search to data
grid_search.fit(x, y)

# Print best hyperparameters
print('Best parameters:', grid_search.best_params_)

# Train XGBoost Regressor with best hyperparameters
best_xgb_reg = xgb.XGBRegressor(n_estimators=grid_search.best_params_['n_estimators'],
                                learning_rate=grid_search.best_params_['learning_rate'],
                                max_depth=grid_search.best_params_['max_depth'])
best_xgb_reg.fit(x, y)

# Make predictions
y_pred = best_xgb_reg.predict(x)

# Print mean squared error
mse = mean_squared_error(y, y_pred)
print('Mean Squared Error:', mse)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

# Load data
#X, y = load_data()

# Define Random Forest Regressor
rf_reg = RandomForestRegressor()

# Define hyperparameters for Grid Search
parameters = {'n_estimators': [100, 500, 1000],
              'max_depth': [3, 6, 9],
              'min_samples_split': [2, 5, 10]}

# Create Grid Search object
grid_search = GridSearchCV(estimator=rf_reg, param_grid=parameters,
                           cv=5, n_jobs=-1, scoring='neg_mean_squared_error')

# Fit Grid Search to data
grid_search.fit(x, y)

# Print best hyperparameters
print('Best parameters:', grid_search.best_params_)

# Train Random Forest Regressor with best hyperparameters
best_rf_reg = RandomForestRegressor(n_estimators=grid_search.best_params_['n_estimators'],
                                     max_depth=grid_search.best_params_['max_depth'],
                                     min_samples_split=grid_search.best_params_['min_samples_split'])
best_rf_reg.fit(x, y)

# Make predictions
y_pred = best_rf_reg.predict(x)

# Print mean squared error
mse = mean_squared_error(y, y_pred)
print('Mean Squared Error:', mse)
```

```python
Accuracy=r2_score(y,y_pred)*100
print(" Accuracy of the model is %.2f" %Accuracy)
```
